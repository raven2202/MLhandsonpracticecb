```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
df=pd.read_csv('carprices.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
dfle=df
dfle['Car Model']=le.fit_transform(dfle['Car Model'])
dfle
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer([('Car Model',OneHotEncoder(drop='first'),[0])],remainder='passthrough')
```

```python
X=ct.fit_transform(dfle.drop('Sell Price($)',axis='columns'))
X
```

```python
Y=dfle['Sell Price($)']
Y
```

```python
re=linear_model.LinearRegression()
re.fit(X,Y)
```

```python
re.predict(pd.DataFrame([[0,1,45000,4]]))
```

```python
re.predict([[1,0,86000,7]])
```

```python
re.score(X,Y)
```
