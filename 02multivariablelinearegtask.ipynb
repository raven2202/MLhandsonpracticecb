```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
df=pd.read_csv("hiring.csv")
df
```

```python
pip install word2number
```

```python
from word2number import w2n
df['experience']=df['experience'].apply(lambda x:w2n.word_to_num(x) if isinstance(x, str) else np.nan)
```

```python
df
```

```python
expmedian=df['experience'].median()
testscoremedian=df['test_score(out of 10)'].median()
```

```python
df['experience']=df.experience.fillna(expmedian)
df['test_score(out of 10)']=df['test_score(out of 10)'].fillna(testscoremedian)
df.rename(columns={'test_score(out of 10)':'test_score', 'salary($)':'salary'},inplace=True)
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['experience','test_score','interview_score(out of 10)']],df.salary)
```

```python
reg.predict(pd.DataFrame({'experience':[2],'test_score':[9],'interview_score(out of 10)':[6]}))
```
