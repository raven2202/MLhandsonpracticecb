```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('homepricesohe.csv')
df
```

```python
dummies=pd.get_dummies(df.town,dtype=int)
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
final=merged.drop(['town','west windsor'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
x=final.drop('price',axis='columns')
x
```

```python
y=final.price
y
```

```python
model.fit(x,y)
```

```python
model.predict(pd.DataFrame([[2800,0,1]]))
```

```python
model.predict(pd.DataFrame([[3200,0,0]]))
```

```python
model.score(x,y)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
dfle=df
dfle.town=le.fit_transform(dfle.town)
dfle
```

```python
X=dfle[['town','area']]
X
```

```python
Y=dfle.price
Y
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(drop='first'), ['town'])  # drop='first' to avoid dummy trap
    ],
    remainder='passthrough'  # Leave other columns (like 'area') as-is
)
```

```python
X_encoded=ct.fit_transform(X)
X_encoded
```

```python
model.fit(X_encoded,Y)
```

```python
model.predict([[1,0,2800]])
```

```python
model.score(X_encoded,Y)
```
